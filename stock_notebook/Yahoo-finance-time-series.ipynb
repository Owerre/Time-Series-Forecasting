## Imports

```python
%matplotlib inline

import pandas_datareader.data as web
import pandas as pd
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf

from datetime import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
mpl.rcParams['font.family'] = 'sans-serif'
sns.set(font_scale = 1.2)
# import plotly 
# import cufflinks as cf
# plotly.tools.set_credentials_file(username='', api_key='')
```

```python
start_date = datetime(2010,1,1) # Start date of the stock prices
end_date = datetime.now() # End date of the stock prices

df = web.DataReader('AAPL', 'yahoo', start_date,end_date) # Apple stock prices
```

```python
df.head() # Display the first five columns
```

```python
df.tail() # Display the last five columns
```

```python
# Plot
df['Close'].plot(figsize = (15,8))

# Label
plt.xlabel('Year')
plt.ylabel('Closing price')
```

```python
rtn = np.log(df/df.shift(1)) # Log return
```

```python
rtn.head()
```

```python
# Plot
rtn['Close'].plot(figsize = (15,8)) # Time plot of log return

# Labels
plt.ylabel('Log-return')
plt.title('Log return of Apple stock prices')
```

```python
plot_acf(df.Close, lags =10)
```
