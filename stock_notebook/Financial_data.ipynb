```python
%matplotlib inline

import pandas_datareader.data as web
import pandas as pd
import numpy as np

from datetime import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
mpl.rcParams['font.family'] = 'sans-serif'
sns.set(font_scale = 1.2)
# import plotly 
# import cufflinks as cf
# plotly.tools.set_credentials_file(username='', api_key='')
```

```python
start_date = datetime(2010,1,1) # Start date of the stock prices
end_date = datetime.now() # End date of the stock prices

names = ['FB','MSFT', 'JPM','AAPL', 'AMZN', 'NFLX', 'SPY'] # Symbol names of companies
df_raw = web.DataReader(names, 'yahoo', start_date,end_date) # Data frame
```

```python
df_raw.head() # Display the first five columns
```

```python
df_raw.tail() # Display the last five columns
```

```python
df = df_raw['Close'] # Select the closing stock price
```

```python
df.info()
```

```python
df = df.dropna() # Drop columns with missing values
```

```python
df.describe().round(2) # Statistical description
```

```python
rtn = np.log(df/df.shift(1)) # Log return
```

```python
rtn.head()
```

```python
rtn['AAPL'].plot(figsize = (15,8)) # Time plot of log return
plt.ylabel('Log-return') # y axis label
plt.title('Log return of Apple stock prices')
```
